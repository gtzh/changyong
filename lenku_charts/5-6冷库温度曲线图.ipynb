```python
import  pandas  as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
from pyecharts.globals import ThemeType
from pyecharts.charts import Line
from pyecharts import options as opts
```

```python
%ls
```

```python
df_lenku_5yue = pd.read_excel('冷库5-6月份温度记录.xlsx',sheet_name='6月份', index_col=0) 
df_lenku_5yue = df_lenku_5yue.iloc[:, 0:5]
df_lenku_5yue.describe()
```

```python

# 折线图
line = Line(init_opts=opts.InitOpts(theme=ThemeType.MACARONS))

# x轴
line.add_xaxis(df_lenku_5yue.index.to_list())

# 每个y轴
line.add_yaxis("冷库温度", df_lenku_5yue["1#冷库（改原料库）"].to_list())


# 图表配置
line.set_global_opts(
    title_opts=opts.TitleOpts(title="6月冷库温度曲线图-1#冷库（改原料库）"),
    tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"),
    yaxis_opts=opts.AxisOpts(min_='dataMin')
)
# 渲染数据
line.render_notebook()
```
