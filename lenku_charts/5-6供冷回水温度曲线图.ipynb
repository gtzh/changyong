```python
import  pandas  as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
df_lengong_5yue = pd.read_excel('低温5-6月放冷供回水温度.xlsx',sheet_name='低温6月份冰水记录', index_col=0) 
df_lengong_5yue = df_lengong_5yue.drop(['Unnamed: 2'], axis=1)
df_lengong_5yue.describe()
```

```python
df_lengong_5yue
```

```python
# 如果没有安装，使用pip install pyecharts安装
from pyecharts.charts import Line, Bar
from pyecharts import options as opts
# 折线图
line = Line()

# x轴
line.add_xaxis(df_lengong_5yue.index.to_list())

# 每个y轴
line.add_yaxis("低温供水温度", df_lengong_5yue["低温供水温度"].to_list())
line.add_yaxis("低温回水温度", df_lengong_5yue["低温回水温度"].to_list())

# 图表配置
line.set_global_opts(
    title_opts=opts.TitleOpts(title="6月供水回水温度曲线图"),
    tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross")
)
# 渲染数据
line.render_notebook()
```

```python
vals_of_supply_water = df_lengong_5yue['低温供水温度'].value_counts().sort_values()
print(vals_of_supply_water)
print('-'*50)
indexs_of_supply_water = vals_of_supply_water.index
vals_of_supply_water = vals_of_supply_water.values
print(type(indexs_of_supply_water))
from pyecharts.charts import Line, Bar

bar = Bar()
bar.add_xaxis(indexs_of_supply_water.to_list())

# 每个y轴
bar.add_yaxis("低温供水温度频次分布", vals_of_supply_water.tolist())

# 图表配置
bar.set_global_opts(
    title_opts=opts.TitleOpts(title="6月供水温度频次图"),
)
# 渲染数据
bar.render_notebook()
```
